```python
import numpy as np
import cv2

img = cv2.imread('/content/gdrive/MyDrive/電腦視覺/圖片/twins.jpg',cv2.IMREAD_GRAYSCALE)

newImg=img[128:256,128:256] #取1/4
newdata=newImg.ravel() # 扁平化
keys=np.unique(newdata) # 去除重複+排序

counts=[]
for key in keys:
  count=np.count_nonzero(newdata==key)
  counts.append(count)
  
print(keys) # x軸
print(counts) # y軸
```

```python
import matplotlib.pyplot as plt
plt.bar(keys, counts, 0.35)
plt.show()
plt.imshow(newImg, cmap='gray') # 1/4圖形
```
