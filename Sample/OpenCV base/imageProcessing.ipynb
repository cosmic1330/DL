```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Blur
image_path = '../images/imageProgess/cat.jpg'
image = cv2.imread(image_path)
image = image[:,:,::-1]
meanBlur = np.ones((5, 5), dtype="float") * (1.0 / (5 * 5))
print(meanBlur)
result = cv2.filter2D(image, -1, meanBlur) # filter2D (src, , , kernel)

plt.figure(figsize=(10,6))
plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
plt.imshow(result)
plt.show()
```

```python
# Sharpen
sharpen = np.array((
	[0, -1, 0],
	[-1, 5, -1],
	[0, -1, 0]), dtype="int")

result = cv2.filter2D(image, -1, sharpen)
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
plt.imshow(result)
plt.show()
```

```python
# Thresholding
img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO) # cv2.threshold(img, , , )
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
plt.imshow(thresh5, cmap='gray')
plt.show()
```

```python
# Edge detection
laplacian = np.array((
	[0, 1, 0],
	[1, -4, 1],
	[0, 1, 0]), dtype="int")

result = cv2.filter2D(image, -1, laplacian)
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
plt.imshow(result)
plt.show()
```

```python
# XSobel
sobelX = np.array((
	[-1, 0, 1],
	[-2, 0, 2],
	[-1, 0, 1]), dtype="int")

result = cv2.filter2D(img, -1, sobelX)
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
plt.imshow(result, cmap='gray')
plt.show()
```

```python
# YSoble
sobelY = np.array((
	[-1, -2, -1],
	[0, 0, 0],
	[1, 2, 1]), dtype="int")

result = cv2.filter2D(img, -1, sobelY)
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
plt.imshow(result, cmap='gray')
plt.show()
```
